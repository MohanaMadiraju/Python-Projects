```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file from Kaggle
data = pd.read_csv('/kaggle/input/sbi-npa-dataset/Risk engagement matrix.csv')

# Extract Likelihood and Impact columns
likelihood = data['Normalized Likelihood']
impact = data['Normalized Impact']
years = data['Year']  # Assuming a Year column exists

# Plot the Risk Engagement Matrix
plt.figure(figsize=(8, 8))
plt.scatter(likelihood, impact, color='blue', s=100, edgecolor='black', label='Risks')

# Add labels for each point
for i, year in enumerate(years):
    plt.text(likelihood[i] + 0.1, impact[i], str(year), fontsize=9)

# Add quadrant reference lines
plt.axhline(3, color='black', linestyle='--', linewidth=0.8)
plt.axvline(3, color='black', linestyle='--', linewidth=0.8)

# Customize the plot
plt.title('Risk Engagement Matrix: SBI NPAs', fontsize=14, fontweight='bold')
plt.xlabel('Likelihood (Low to High)', fontsize=12)
plt.ylabel('Impact (Low to High)', fontsize=12)
plt.xticks(range(1, 6))
plt.yticks(range(1, 6))
plt.grid(which='major', linestyle='--', linewidth=0.5, color='gray')

# Add legend
plt.legend(['Risks'], loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
```
